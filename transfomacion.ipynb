```python
#Danny Rodriguez
import speech_recognition as sr
from pydub import AudioSegment

# Definir la función para transcribir el audio
def transcribe_audio(file_path):
    # Inicializar el reconocedor de audio
    recognizer = sr.Recognizer()
    
    # Cargar el archivo de audio
    audio = AudioSegment.from_wav(file_path)
    
    # Dividir el audio en segmentos de 1 minuto (60000 milisegundos) para evitar problemas con archivos grandes
    audio_chunks = [audio[i:i+60000] for i in range(0, len(audio), 60000)]
    
    # Transcribir cada segmento
    transcription = ""
    for i, chunk in enumerate(audio_chunks):
        chunk.export(f"chunk{i}.wav", format="wav")
        with sr.AudioFile(f"chunk{i}.wav") as source:
            audio_listened = recognizer.record(source)
            try:
                text = recognizer.recognize_google(audio_listened, language="es-ES")
                transcription += text + " "
            except sr.UnknownValueError:
                print(f"Google Speech Recognition could not understand chunk {i}")
            except sr.RequestError as e:
                print(f"Could not request results from Google Speech Recognition service; {e}")
    
    return transcription

# Definir la ruta al archivo de audio
audio_file_path = "C:/Users/Usuario/Desktop/Pasantias/Actividades/Junio/04/audio.wav"

# Llamar a la función y obtener la transcripción
transcription = transcribe_audio(audio_file_path)

# Guardar la transcripción en un archivo de texto
with open("transcription.txt", "w", encoding="utf-8") as file:
    file.write(transcription)

# Mostrar la transcripción
print(transcription)
```
